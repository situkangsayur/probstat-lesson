```python
import pandas as pd

df = pd.read_csv('PlayTennis.csv')
df
```

```python
display(df.head()) # View the first few rows
display(df.tail())
display(df.info()) # Get information about data types and non-null values
display(df.describe()) # Get descriptive statistics for numerical columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in ['Outlook', 'Temperature', 'Humidity', 'Wind', 'Play Tennis']:
    df[column + '_encoded'] = le.fit_transform(df[column])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
display(df)

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Outlook', hue='Play Tennis')
plt.title('Play Tennis based on Outlook')
plt.show()
```
